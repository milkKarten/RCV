```
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib.lines import Line2D   
import numpy as np 

# Load simple house
house = [[0,0,0],#1: 2,3,4
         [1,0,0],#2: 5,6
         [0,1,0],#3: 5,7
         [0,0,1],#4: 6,7,9
         [1,1,0],#5: 8
         [1,0,1],#6: 8,9
         [0,1,1],#7: 8,9
         [1,1,1],#8: 9
         [0.5,0.5,1.5]]  #9:      

# Calibration parameters
M_ext = np.matrix([[0.707, 0.707, 0, -3],
                   [-0.707,  0.707, 0, -0.5],
                   [0,0,1,3]])
M_int = np.matrix([[-100,0,200],
                   [0,-100,200],
                   [0,0,1]])
M = M_int*M_ext
p = []
for point in house:
  point.append(1)
  P = np.array(point)
  p.append(np.dot(M, P))

fig = plt.figure()
ax = fig.add_subplot(111)
out_x = [0]
out_y = [0]
for point in p:
  z = point[0,2]
  x_new = point[0,0]/z
  y_new = point[0,1]/z
  out_x.append(x_new)
  out_y.append(y_new)
ax.axis([230,310,180,250])

line = Line2D([out_x[1],out_x[2]], [out_y[1],out_y[2]])
ax.add_line(line)
line = Line2D([out_x[1],out_x[3]], [out_y[1],out_y[3]])
ax.add_line(line)
line = Line2D([out_x[1],out_x[4]], [out_y[1],out_y[4]])
ax.add_line(line)

line = Line2D([out_x[2],out_x[5]], [out_y[2],out_y[5]])
ax.add_line(line)
line = Line2D([out_x[2],out_x[6]], [out_y[2],out_y[6]])
ax.add_line(line)

line = Line2D([out_x[3],out_x[5]], [out_y[3],out_y[5]])
ax.add_line(line)
line = Line2D([out_x[3],out_x[7]], [out_y[3],out_y[7]])
ax.add_line(line)

line = Line2D([out_x[4],out_x[6]], [out_y[4],out_y[6]])
ax.add_line(line)
line = Line2D([out_x[4],out_x[7]], [out_y[4],out_y[7]])
ax.add_line(line)
line = Line2D([out_x[4],out_x[9]], [out_y[4],out_y[9]])
ax.add_line(line)

line = Line2D([out_x[5],out_x[8]], [out_y[5],out_y[8]])
ax.add_line(line)

line = Line2D([out_x[6],out_x[8]], [out_y[6],out_y[8]])
ax.add_line(line)
line = Line2D([out_x[6],out_x[9]], [out_y[6],out_y[9]])
ax.add_line(line)

line = Line2D([out_x[7],out_x[8]], [out_y[7],out_y[8]])
ax.add_line(line)
line = Line2D([out_x[7],out_x[9]], [out_y[7],out_y[9]])
ax.add_line(line)

line = Line2D([out_x[8],out_x[9]], [out_y[8],out_y[9]])
ax.add_line(line)
plt.show()
```
